```python
import pandas as pd

data = [
    ['Sunny', 'Warm', 'Normal', 'Strong', 'Warm', 'Same', 'Yes'],
    ['Sunny', 'Warm', 'High',   'Strong', 'Warm', 'Same', 'Yes'],
    ['Rainy', 'Cold', 'High',   'Strong', 'Warm', 'Change', 'No'],
    ['Sunny', 'Warm', 'High',   'Strong', 'Cool', 'Change', 'Yes']
]

columns = ['Sky', 'AirTemp', 'Humidity', 'Wind', 'Water', 'Forecast', 'EnjoySport']
df = pd.DataFrame(data, columns=columns)
display(df)

def find_s(dataset, target_col):
    positive_examples = dataset[dataset[target_col] == "Yes"]
    hypothesis = list(positive_examples.iloc[0][:-1])
    for index, row in positive_examples.iterrows():
        example = list(row[:-1])
        for i in range(len(hypothesis)):
            if example[i] != hypothesis[i]:
                hypothesis[i] = "?"
    return hypothesis

hypothesis = find_s(df, 'EnjoySport')
print("\nFinal Hypothesis learned by Find-S:")
print(hypothesis)
```
